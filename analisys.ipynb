```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def parseString(s:str):
    name, *data = s.split(",")
    particle, dim, pos = name.split("_")
    data = [int(x) for x in data[:-1]]
    res = {"particle": particle,
           "dim": dim,
           "pos": float(pos) if len(pos)>0 else None,
           "data": np.array(data)}

    return res

```

```python
with open('data/data.csv') as f:
    stat = [parseString(s) for s in f.read().split("\n")[:-1]]

stat[0]
```

```python
e_stat_x = np.array([[st['pos'], st['data'].mean()] for st in stat if st['particle']=='e-' and st['dim']=='x'])
e_stat_y = np.array([[st['pos'], st['data'].mean()] for st in stat if st['particle']=='e-' and st['dim']=='y'])
e_stat_z = np.array([[st['pos'], st['data'].mean()] for st in stat if st['particle']=='e-' and st['dim']=='z'])
e_stat_uniform = ([st['data'] for st in stat if st['particle']=='e-' and st['dim']=='uniform'])[0]

fig,ax = plt.subplots(2,2)
fig.set_size_inches(9,9)

fig.suptitle(r"$e^- (E = 1 MeV)$")

ax[0][0].plot(e_stat_x[:,0], e_stat_x[:,1])
ax[0][0].set_xlabel('x')
ax[0][0].set_ylabel(r'$<N_{photons}>$')
ax[0][0].grid()

ax[0][1].plot(e_stat_y[:,0], e_stat_y[:,1])
ax[0][1].set_xlabel('y')
ax[0][1].set_ylabel(r'$<N_{photons}>$')
ax[0][1].grid()

ax[1][0].plot(e_stat_z[:,0], e_stat_z[:,1])
ax[1][0].set_xlabel('z')
ax[1][0].set_ylabel(r'$<N_{photons}>$')
ax[1][0].grid()

ax[1][1].hist(e_stat_uniform)
ax[1][1].set_xlabel(r'$N_{photons}$')
ax[1][1].grid()
```
